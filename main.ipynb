```python
import requests
import pandas as pd

url = 'https://webservice.recruit.co.jp/hotpepper/gourmet/v1/'

start = 1
count = 100

params = {
    'key':'ba58afbfbe75f488',
    'keyword':'五反田駅',
    'format':'json',
    'start':start,
    'count':count
}

res = requests.get(url, params)

shops = res.json()['results']['shop']
shop = shops[0]

datum = {
    'name':shop['name'],
    'address':shop['address'],
    'capacity':shop['capacity'],
    'access':shop['access'],
    'open':shop['open'],
    'budget':shop['budget']['name'],
    'urls':shop['urls']['pc']
}
datum
data = []
for shop in shops:
    datum = {
        'name':shop['name'],
        'address':shop['address'],
        'capacity':shop['capacity'],
        'access':shop['access'],
        'open':shop['open'],
        'budget':shop['budget']['name'],
        'urls':shop['urls']['pc']
    }
    data.append(datum)
data[:5]
df = pd.DataFrame(data)

_budget = df['budget'][0]
int(_budget.split("～")[1].replace("円",""))
lower_budgets = []
upper_budgets = []

for _budget in df["budget"]:
    if _budget != "":
        lower_budget = int(_budget.split('～')[0])
        upper_budget = int(_budget.split("～")[1].replace("円",""))
    else:
        lower_budget = 0
        upper_budget = 0
    lower_budgets.append(lower_budget)
    upper_budgets.append(upper_budget)
df['lower_budget'] = lower_budgets
df['upper_budget'] = upper_budgets
df_filtered = df[df['upper_budget'] <= 4000]
df_filtered.head()
```
